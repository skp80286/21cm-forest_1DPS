```python
import glob
import numpy as np

test_points = [[-3,0.11], [-3,0.80], [-1,0.11], [-1,0.80], [-2,0.52]]
filepattern = str('%sF21_noisy_21cmFAST_200Mpc_z%.1f_fX%s_xHI%s_%s_%dkHz_t%dh_Smin%.1fmJy_alphaR%.2f.dat' % 
               ('../data/21cmFAST_los/F21_noisy/', 6,'*', '*', 'uGMRT', 8, 500, 64.2, -0.44))
print(f"Loading files with pattern {filepattern}")

datafiles = glob.glob(filepattern)
datafiles = sorted(datafiles)
print(f"found {len(datafiles)} files")

train_points = []  # Initialize an empty list to store training points

for file in datafiles:
    xHI_str = file.split('xHI')[-1].split('_')[0]  # Get the value after 'xHI=' and before the next '_'
    logfX_str = file.split('fX')[-1].split('_')[0]  # Get the value after 'fX=' and before the next '_'
    
    xHI = float(xHI_str)
    logfX = float(logfX_str)
    if [xHI, logfX] not in test_points:
        train_points.append([logfX, xHI])  # Add to train_points if not present

train_points = np.array(train_points)
test_points = np.array(test_points)
base.initplt()
# Create a figure and axes object
fig, ax = plt.subplots(figsize=(10, 6))

# Create a joint plot to show the distribution and concentration
#ax.set_title('Distribution of xHI and logfX in training and test data')
ax.set_title(r'$z=6$')
ax.set_xlabel(r'$\langle x_{HI}\rangle$')
ax.set_ylabel(r'$log_{10} f_X$')
ax.minorticks_on()
# Customize major ticks
ax.tick_params(
        which='major',
        direction='in',
        length=10,
        width=1,
        #labelsize=12,
        top=True,
        bottom=True,
        left=True,
        right=True
    )
# Customize minor ticks
ax.tick_params(
        which='minor',
        direction='in',
        length=5,
        width=1,
        top=True,
        bottom=True,
        left=True,
        right=True
    )#plt.colorbar(label='Density', ax=ax)
ax.scatter(train_points[:,1], train_points[:,0], label="Training point", s=25, alpha=0.7, edgecolor='blue')

ax.scatter(test_points[:,1], test_points[:,0], marker='*', color='red', label="Test point", s=196, edgecolor='red', linewidth=.1)

plt.legend(fontsize=12)
plt.savefig("tmp_out/training_test_points.pdf", format="pdf", dpi=300, bbox_inches='tight')
plt.show()
    
```
